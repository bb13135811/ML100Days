# 作業 

### 1. 根據題目給的 DataFrame 完成下列操作：

```
- 計算每個不同種類 animal 的 age 的平均數
- 計算每個不同種類 animal 的 age 的平均數
- 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
- 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
```


````python
# 記得先 Import 正確的套件

import pandas as pd
import numpy as np
````

````python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)

df
````

````python
# 1.計算每個不同種類 animal 的 age 的平均數

print("mean of cat's age:", df.loc[df['animal'] == 'cat'].age.mean())
print("mean of cat's age:", df.loc[df['animal'] == 'dog'].age.mean())
print("mean of cat's age:", df.loc[df['animal'] == 'snake'].age.mean())
````

````python
# 2.將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序

df.sort_values(by= ['age', 'visits'], ascending= [True, False])
````

````python
# 3.將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False

df.replace({'yes': True, 'no': False}, inplace=True)
df
````

### 2. 一個包含兩個欄位的 DataFrame，將每個數字減去

* 1) 該欄位的平均數
* 2) 該筆資料平均數

````python
df = pd.DataFrame(np.random.random(size=(5, 3)))
df
````

````python
a = df.iloc[:, 0].mean()
b = df.iloc[:, 1].mean()
c = df.iloc[:, 2].mean()

df.sub(pd.Series([a, b, c]),axis='columns')
````

````python
mean = sum(df.sum()) / sum(df.count())
df = df- mean
df
````

### 3. 承上題，請問：

1) 哪一比的資料總合最小
2) 哪一欄位的資料總合最小
